## SDSS Analysis Part 1

Here is some info about the selected 50,000-item subset of the SDSS:

```python
import pandas as pd

df = pd.read_csv('SDSS_500k_extended_v2.csv')

print(df.info())
print("""


""")
```

Now, a small 5% portion of the subset will be placed aside as a test set to validate the accuracy of the ML model:

```python
from sklearn.model_selection import train_test_split

stratified_train_set, stratified_test_set = train_test_split(df, test_size=0.05, stratify=df["class"], random_state=67)

print(stratified_train_set["class"].value_counts() / len(stratified_train_set))
print("")
print(stratified_test_set["class"].value_counts() / len(stratified_test_set))
```
