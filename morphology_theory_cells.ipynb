### Galaxy Morphology — Shape as a Window into History

---

<div style="width: 1200px">

Every galaxy has a shape. And that shape is not arbitrary — it is the accumulated record of billions of years of collisions, mergers, gas inflows, star formation, and quenching. Learning to read a galaxy's morphology is like learning to read a face: with practice, a single glance can tell you a great deal about its past.

The systematic classification of galaxy shapes began in the 1920s with Edwin Hubble, who organised the bewildering variety of galaxy forms into a simple diagram that has become one of the most iconic images in astronomy.

</div>

---

### The Hubble Tuning Fork

---

<div style="width: 1200px">

In 1926, Edwin Hubble published a classification scheme that arranged galaxies into a fork-shaped diagram — now universally known as the **Hubble Tuning Fork**. It organises galaxies along two axes simultaneously:

**Left to right — from smooth to structured.** Moving from left to right along the diagram, galaxies develop progressively more disc structure, more prominent spiral arms, and smaller central bulges. The leftmost galaxies are entirely smooth and featureless; the rightmost are wispy, arm-dominated systems.

**The fork split — barred vs unbarred.** Where the diagram forks, it separates galaxies with a straight central bar of stars (barred spirals, bottom arm) from those without one (normal spirals, top arm). About half of all disc galaxies in the local universe have bars — the Milky Way itself is thought to be a barred spiral.

The classes, moving from left to right, are:

| Class | Name | Key features |
|-------|------|--------------|
| E0–E7 | Ellipticals | Smooth, round to elongated, no disc, no arms. The number gives the ellipticity: E0 is circular, E7 is highly flattened |
| S0 | Lenticular | Has a disc but no spiral arms — the bridge between ellipticals and spirals |
| Sa, Sb, Sc | Spirals | Disc with spiral arms; moving from Sa→Sc, the bulge shrinks and arms become more open and patchy |
| SBa, SBb, SBc | Barred spirals | Like Sa–Sc but with a straight bar of stars through the centre, from which the arms emerge |

Hubble originally called ellipticals "early-type" and spirals "late-type" — implying an evolutionary sequence from left to right. We now know this is **wrong**: galaxies do not evolve along the tuning fork from elliptical to spiral. If anything, the direction is reversed — spirals can be transformed into ellipticals through mergers, but not the other way around. The terminology "early" and "late" type has unfortunately stuck, and you will still see it used in the literature today.

</div>

* The **gas fraction** tells a similar story — ellipticals are almost gas-free (their gas was consumed or expelled long ago), while late spirals are gas-rich and still actively building stars.

These are not coincidences. They reflect a common underlying physical story: galaxies on the left have finished their main episode of star formation, while galaxies on the right are still in the middle of it.

### Beyond the Tuning Fork — Irregulars and Mergers

---

<div style="width: 1200px">

The Hubble classification covers the majority of nearby galaxies, but it misses an important minority: **irregular galaxies** that don't fit neatly into any category.

Irregular galaxies (type Irr) are typically small, asymmetric, and intensely star-forming. The Large and Small Magellanic Clouds — the Milky Way's nearest large neighbours, visible from the Southern Hemisphere as faint patches of the night sky — are classic examples. Their chaotic shapes are often caused by tidal interactions with a larger neighbour.

At higher redshifts (earlier cosmic times), irregular and disturbed morphologies become much more common, because the universe was denser and galaxy collisions were more frequent. This is one of the key pieces of evidence that galaxies build up their mass through mergers over cosmic time.

**Galaxy mergers** deserve special mention because they can completely transform a galaxy's morphology in a relatively short time (~1 billion years). When two spiral galaxies of similar mass collide — a **major merger** — the result is typically an elliptical galaxy. The disc structures are destroyed in the violent gravitational interaction; the stellar orbits are randomised; and a burst of star formation is triggered by gas compression before the gas is finally exhausted. This "merger-to-elliptical" pathway is one of the primary ways the red sequence grows.

</div>

### Connecting Morphology to our SDSS Measurements

---

<div style="width: 1200px">

We don't have visual morphology classifications for every galaxy in our 500,000-object SDSS sample — visually classifying that many galaxies would take years, even with volunteer projects like Galaxy Zoo. Instead, we use the quantitative structural measurements in the dataset as **morphological proxies**:

| SDSS measurement | What it proxies | Tuning fork connection |
|-----------------|-----------------|------------------------|
| `g - r` colour | Star formation rate / stellar age | Redder → further left (ellipticals) |
| Concentration $C$ | Bulge dominance | Higher $C$ → further left |
| `fracDeV_r` | Sérsic index / profile shape | Higher → de Vaucouleurs (elliptical) |
| `petroRad_r` | Physical size | Larger → more extended disc |
| `expAB_r` | Axis ratio (roundness) | Lower → more elongated / edge-on disc |

None of these individually is a perfect morphological classifier — a round galaxy with low $C$ could be either a face-on spiral or a true elliptical. But in combination, they paint a coherent picture. When we see a galaxy with high $C$, high `fracDeV_r`, red $g - r$, and small `petroRad_r`, we can be fairly confident it's an elliptical on the left of the tuning fork. When we see low $C$, low `fracDeV_r`, blue $g - r$, and large `petroRad_r`, we're almost certainly looking at a late spiral.

This is what our plots in this notebook have been doing all along — not just showing numbers, but reading the morphological story encoded in the data.
